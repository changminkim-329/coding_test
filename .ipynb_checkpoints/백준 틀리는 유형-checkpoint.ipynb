# 백준 시간초과 

백준 오답 중 가장 그지 같은 유형. 답은 맞는데, 해당 알고리즘이 정답 알고리즘 보다 불 필요한 처리를 더해서 속도가 느림

백준 같은 경우 시간초과가 일어나면 python 대신 pypy3를 사용하면 해결 되는 경우가 있다.

### 백준 특정 경우에 대해서 틀림

특정 경우에 대해서는 맞지만, 특정 경우에 대해서 틀린다. 

어려운 이유: 어떤 경우에 대해서 틀렸는 지 파악하기 어려움

but 해당 경우를 알 경우 빠르게 문제 해결 가능


접근 방법

1. 일단 다른 문제를 풀고 남은 시간에 풀기 (전제 조건: 내 기준에서는 분명히 맞는 거 같은데)
- 코테도 물론 정답을 많이 맞출 수 록 좋다. 하지만 대부분은 코테는 정답 보단 어떻게 문제를 접근 했는지에 대한 문제 풀이 위주로 보기 때문에 다른 문제로 넘어 가는 것이 좋다.(물론 특정 경우에 대해서만 틀리 다는 것을 인지해야 됨)

2. 문제에 나와 있는 범위에 대해서 for 를 돌려 특정 경우에 대해서 틀리는지 확인 (시간이 남았을 때)
- 대부분 앞에서 문제를 찾을 수 있음

#### ex) 10250번 문제
- (1 ≤ H, W ≤ 99, 1 ≤ N ≤ H × W). 

```python
T = int(input())
for i in range(T):
    H,W,N = map(int,input().split()) #H: 높이, W: 거리, N: 손님

    if (N//H+1) < 10:
        N_ = '0'+str(N//H+1)
    else:
        N_ = str(N//H+1)
    print(str(N%H)+N_)
    print(int(str(N%H)+N_))
```

해당 문제의 입력에 대해 for문을 사용해 모든 경우를 일일이 파악

```python
for H in range(1,99+1):
    for W in range(1,99+1):
        for N in range(1,H*W+1):
            #H,W,N = map(int,input().split()) #H: 높이, W: 거리, N: 손님

            if (N//H+1) < 10:
                N_ = '0'+str(N//H+1)
            else:
                N_ = str(N//H+1)
            print(H,W,N,":",str(N%H)+N_)
```

```python
for H in range(1,99+1):
    for W in range(1,99+1):
        for N in range(1,H*W+1):
            #H,W,N = map(int,input().split()) #H: 높이, W: 거리, N: 손님

            if N%H == 0:
                H_ = str(H)
            else :
                H_ =str(N%H)

            if (N//H+1) < 10:
                N_ = '0'+str(N//H+1)
            else:
                N_ = str(N//H+1)
            print(H,W,N,":",H_+N_)
```

```python
for H in range(1,99+1):
    for W in range(1,99+1):
        for N in range(1,H*W+1):
            #H,W,N = map(int,input().split()) #H: 높이, W: 거리, N: 손님

            if N%H == 0:
                H_ = str(H)
                N_ = N//H
            else :
                H_ =str(N%H)
                N_ = N//H+1
            
            if (N//H+1) < 10:
                N_ = '0'+str(N_)
            else:
                N_ = str(N_)
            print(H,W,N,":",H_+N_)
```
